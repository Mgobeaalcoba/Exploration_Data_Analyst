# Llamar los datos 

```
#Llamamos el set de datos público 
import pandas as pd
url_wine_red = 'https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/winequality-red.csv'
url_wine_white = 'https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/winequality-white.csv'

```

```
red=pd.read_csv(url_wine_red,delimiter=';')
white=pd.read_csv(url_wine_white,delimiter=';')
```

```
red.head()
```

```
white.head()
```

```
red.append(white)
```

```
# Genero una nueva columna donde voy a detallar que tipo de vino se trata
red['color']='red'
white['color']='white'
```

```
red.append(white) # Primera forma de unir usando el append
```

```
pd.concat([white,red]) # Segunda forma de unir usando el concat de pandas
```

Tmb podemos usar el merge. Para uniones del tipo "left", "right", "union", "join", etc como si estuviesemos trabajando con SQL

```
left=pd.DataFrame({'Key':['key1', 'key2', 'key3']
              ,'A':['a1', 'a2', 'a3']
})
```

```
right=pd.DataFrame({'Key':['key3', 'key4', 'key5']
              ,'B':['b1', 'b2', 'b3']
})
```

```
left
```

```
right
```

```
pd.merge(left, right,on='Key') # merge de tipo join, es el que sale por default. 
```

```
pd.merge(left, right,on='Key',how='left') # Si especifico el how entonces puedo decirle que
# haga una union por izquierda, por derecha, etc.
```
